```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
import seaborn as sn
```

```python
df = pd.read_csv("D:\Portfolio\GroceryBasketAssociation\Groceries_dataset.csv")
df.head()
```

```python
#find null size
total_cell  = np.product(df.shape)
null_cell = df.isnull().sum().sum()
null_percent = (null_cell)/(total_cell)*100
print("total cells = ",total_cell)
print("null cells = ",null_cell)
print("%null cells = ",null_percent)
```

```python
#Build purchased basket per each customer
group_by = df.groupby(['Member_number','itemDescription'])['itemDescription'].count().unstack().reset_index().fillna(0).set_index('Member_number')
group_by = group_by.where(group_by<1,1)
group_by
```

```python
#Find most frequent products
from mlxtend.frequent_patterns import apriori, association_rules
frequent_itemsets = apriori(group_by, min_support = 0.03, use_colnames =True)
frequent_itemsets = frequent_itemsets.sort_values('support',ascending= False)
frequent_itemsets
```

```python
#Find associations
rules = association_rules(frequent_itemsets,metric= 'confidence',min_threshold=0.3)
rules.sort_values(by='confidence', ascending = False).head(10)
```

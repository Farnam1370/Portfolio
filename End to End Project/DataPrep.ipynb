```python
import pandas as pd

```

```python
df = pd.read_csv("D:\Portfolio\WebScraping\Pararius\pararius_dataset.csv")
```

```python
df.head()
```

```python
#Extract numbers

df.price = df.price.str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int)
df.area = df.area.str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int)
df.room = df.room.str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int)
```

```python
df.head()
```

```python
# Extract zipcode elements

df[['zip_num','zip_char','city','district']] = df.zipcode.str.split(' ',3,expand=True)
df.district = df.district.str.replace(')','')
df.district = df.district.str.replace('(','')
```

```python
# Make standard zipcode

df['zip'] = df.zip_num + df.zip_char
df.head()
```

```python
# Make property type and address columns

df[['type','address']] = df.title.str.split(' ',1,expand=True)
df.head()
```

```python
# Build new data frame with correct columns

new_df = df[['title','type','interior','address','zip','city','district','area','room','price']]
new_df.columns = ['title','type','interior','address','zipcode','city','district','area_square_meter','room','monthly_price_euro']
```

```python
# Remove duplicates

new_df.drop_duplicates(subset= ['zipcode','address'],keep='first',inplace=True)
new_df.head()
```

```python
pd.DataFrame.to_csv(new_df,"D:\Portfolio\WebScraping\Pararius\pararius_refined.csv",index=False)
```
